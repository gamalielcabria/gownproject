````python
# utility_code.py:
import base64
import re

"""
Utility code fetch from Jie Jenn's Image analysis With OpenAI API GPT4 Vision In Python
(https://www.youtube.com/watch?v=_1ujhANv6a4)
"""


def encode_image(image_path):
    """
    Encodes and image to base 64 String
    """
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

def extract_code(text):
    # Pattern to match the code block
    pattern = r'```python\s+(.*?)\s+```'

    #extract the code
    match = re.search(pattern, text, re.DOTALL)
    if match:
        return match.group(1)
    return None

def extract_details(text):
    pattern_one = r'```(.*?)```'
    match = re.search(pattern_one,text, re.DOTALL)
    if match:
        details = match.group(1)
        return details
    return None
````

## Analysing the GC Results
### Processing the dataframe
We process the save CSV from previous block.

````python
import pandas as pd
import re
from datetime import datetime, timedelta
filepath = f"{output_file}" #Chained from previous block
df = pd.read_csv(filepath)
#print(df)

# Setting up functions for later
def determine_type(filename):
    if 'CALCURVE' in filename:
        return 'CALIBRATION'
    elif 'SMPL' in filename:
        return 'SAMPLE'
    elif 'CTRL' in filename:
        return 'CONTROL'
    else:
        return 'OTHER'

def extract_info(row):
    if row['Type'] in ['SAMPLE', 'CONTROL']:
        parts = row['File'].split('-')
        if len(parts) >= 3:
            return parts[0], parts[1]
    return None, None

def determine_start(sample):
    if sample and str(sample) != "None":    
        start_time = {
            'S1': '1100PM',
            'S2': '1100AM',
            'S3': '1100AM',
            'S4': '1100AM',
        }
        return start_time.get(sample,'723PM')
    return None

def parse_time(time_str):
    if time_str is None:
        return None
    return datetime.strptime(time_str, "%I%M%p")

def parse_date(date_str, format):
    return datetime.strptime(date_str, format)

# Combine date and time into full datetime objects
def combine_date_time(date, time):
    if pd.notna(date) and pd.notna(time):
        return datetime.combine(date.date(), time.time())
    return None

# Main analysis technically
df_split = df['Filename'].str.split(r'_', expand=True)
df2 = df[['RetTime','Area','Amount']]
df2['Date'] = df_split[1]
df2['Time'] = df_split[2]
df2['File'] = df_split[3].apply(lambda x: re.sub(r'DE-|.D','',x))
df2['Type'] = df_split[3].apply(determine_type)

df2[['Sample', 'Start_Date']] = df2.apply(lambda row: pd.Series(extract_info(row)), axis=1)
df2['Start_Time'] = df2['Sample'].apply(determine_start)

# Parse Date and Start_Date
df2['Parsed_Date'] = df2['Date'].apply(lambda x: parse_date(x, "%Y%m%d"))
df2['Parsed_Start_Date'] = df2['Start_Date'].apply(lambda x: parse_date(x, "%m%d%Y") if pd.notna(x) else None)

# Parse Time and Start_Time
df2['Parsed_Time'] = df2['Time'].apply(parse_time)
df2['Parsed_Start_Time'] = df2['Start_Time'].apply(parse_time)

# Combine Date and Time into full datetime
df2['Full_DateTime'] = df2.apply(lambda row: combine_date_time(row['Parsed_Date'], row['Parsed_Time']), axis=1)
df2['Full_Start_DateTime'] = df2.apply(lambda row: combine_date_time(row['Parsed_Start_Date'], row['Parsed_Start_Time']), axis=1)

# Calculate the total difference in hours
df2['Total_Difference_Hours'] = df2.apply(
    lambda row: (row['Full_DateTime'] - row['Full_Start_DateTime']).total_seconds() / 3600 
                if pd.notna(row['Full_DateTime']) and pd.notna(row['Full_Start_DateTime']) else None,
    axis=1
)

# Drop temporary parsed columns if not needed
df2.drop(['Time', 'Start_Time', 'Date', 'Start_Date','Parsed_Time', 'Parsed_Start_Time', 'Parsed_Date', 'Parsed_Start_Date'], axis=1, inplace=True)

df2
````

````python
import matplotlib.pyplot as plt

# Replace '-' with 0 in the Amount column and ensure it is numeric
df2['Amount'] = df2['Amount'].replace('-', 0)
df2['Amount'] = pd.to_numeric(df2['Amount'], errors='coerce')

# Filter the DataFrame
filtered_df2 = df2[
    (df2['Sample'].notna()) & 
    (~df2['Type'].isin(['CALIBRATION', 'CONTROL']))
]

# Check for any invalid Total_Difference_Hours or Amount values and drop them
filtered_df2 = filtered_df2.dropna(subset=['Total_Difference_Hours', 'Amount'])

filtered_df2 = filtered_df2.sort_values(by=['Sample', 'Total_Difference_Hours']).reset_index(drop=True)

# Define line styles for different groups
line_styles = {
    'S': '-',    # Solid line for S-prefixed samples
    'A': '--',   # Dashed line for A-prefixed samples
    'B': '-.',   # Dash-dot line for B-prefixed samples
    'C': ':',    # Dotted line for C-prefixed samples
}

# Create a line plot for each Sample value
plt.figure(figsize=(10, 6))

for sample, group in filtered_df2.groupby('Sample'):
    # Determine line style based on the prefix of the sample
    line_style = line_styles.get(sample[0], '-')  # Default to solid line if prefix is unknown
    plt.plot(group['Total_Difference_Hours'], group['Amount'], label=f'Sample: {sample}', linestyle=line_style)

# Customize the plot
plt.title('Amount vs. Total Incubation in Hours by Sample', fontsize=14)
plt.xlabel('Total Incubation in Hours', fontsize=12)
plt.ylabel('Amount', fontsize=12)
plt.legend(title='Sample', fontsize=10)
plt.grid(True)

plt.figtext(
    0.5, -0.1,
    "Solid line: Groundwater Mix Inocula (prefix 'S')\nDashed line: Old Denitrification Samples (prefix 'A', 'B', 'C')",
    wrap=True, horizontalalignment='center', fontsize=10, color='gray'
)

# Show the plot
plt.show()
````

## Performing Cell Counting
### ImageJ Protocol

1. DAPI image capture:
    - DAPI images are manually captured with improved neubaeur heamocytometer. DAPI filter at 25% source light intensity at 10X unless stated
        - The improved neubaeur heamocytometer is composed of 9 large boxes with 1mm with
        - Each `1mm box` is subdivided into 16 with sides having size of `0.25mm/250micron`
        - The middle grid has minor grids within this 16 with side having `0.050mm or 50 microns`
    - Image captured at `40s-60s` capture time
    - Curves adjusted before saving

2. Image Processing:
    - Import File
      - `File > Open`
    
    - Duplicate image
      - `Image > Duplicate`
    
    - Change Image Type
      - `Image > Type > 8-bit`
        
    - Removing the dead pixels
      - `Process > Filters > Median`
      - Set `Radius` to 2.0  

    - Evening the contrast across the image
      - Flatten background  using `Rolling Ball Algorithm`
      - Go to `Processs > Subtract Background` and set `Rolling Ball Radius` to 50
      - Tick on the **Sliding Paraboloid** before applying

    - Enhance Contrast:
      - `Process > Enhance Contrast`
      - Set `Saturated Pixels` to 2.0%
    
    - Set the scale for using the haemocytometer
      - Using `Straight` tool, draw a line on the grid of haemocytometer
      - In `Analyze > Set Scale`, set `Known distance` to 250 microns if using 10X or 50 microns if using 40x
      - Set `Unit of length` to microns
      - ***NOTE***: Using another image of a blank haemocytometer as a guide, cell can be counted even without the grid by using the standard scale of 3.5642pixels/microns
        
    - Adjust Threshold
      - Manually adjust the threshold to see the last visible cell to be selected in the threshold
      - Set the threshold to (29,255) as default in the macros
      - Press `Apply`
      - `Process > Binary > Dilate`  to add pixels
      - Use `Process > Binary > Watershed` top automate separation of ImageJ assumed cells
    
    - Save the file as a different image: `<Old Prefix>_edited.tif`
    
    
2. Cell Counting:
    - Analyze the cell
      - Using the `Rectangle` selection tool, target one grid shown in the haemocytometer
      - Set the `Size (microns^ 2)` to 8-Infinity
      - Set `Circularity` to 0.3-1.0
      - Tick the following:
        - Display Results
        - Exclude on Edges
        - Include Holes
        - Summarize
        - Add to Manager
      - Run 
      - Save the Overlay
      - **GUI ONLY**: Close Results windows except for Summary 
      - Repeat the ***Analyze the Cell*** process to other target grid box
    
    -Saving the image for training with AI
      - Save the Summary if finished with the image
      - Flatten the Overlay:
          - Change the color using: `Image > Overlay > Overlay Options` and apply it
          - Go to `Image > Overlay > Flatten`
      - Save the file as a different image: `<Old Prefix>_overlay.tif`
